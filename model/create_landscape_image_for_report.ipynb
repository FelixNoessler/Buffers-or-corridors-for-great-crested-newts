```python
import matplotlib.colors as mat_color
import create_landscape as cl
%pylab inline
```

```python
is_center_of_pond, pond, pond_id = cl.create_landscape(no_of_circles=7, radius=15)
corridors = cl.make_corridors(is_center_of_pond, pond)
buffers = cl.make_buffers(corridors, is_center_of_pond)

buffers[buffers == 35] = 1
buffers[buffers == 55] = 2
buffers[buffers == 105] = 3

corridors[corridors == 35] = 1
corridors[corridors == 55] = 2
corridors[corridors == 105] = 3

manual_cmap = mat_color.ListedColormap(['brown', 'cornsilk', 'blue'])
```

```python
plt.figure(figsize=(16,8))

plt.subplot(121)
plt.imshow(buffers, cmap=manual_cmap, origin='lower', interpolation='nearest')
plt.title('Buffer scenario', size=22)
plt.text(100, 200, 'Matrix:\nAgricultural field', size=18, ha='center')
plt.text(220, 290, 'Pond', size=18, ha='center', color='blue')
plt.text(275, 250, 'Buffer:\nWoodland', size=18, ha='center', color='brown')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.subplot(122)
plt.imshow(corridors, cmap=manual_cmap, origin='lower', interpolation='nearest')
plt.title('Corridor scenario', size=22)
plt.text(100, 200, 'Matrix:\nAgricultural field', size=18, ha='center')
plt.text(220, 280, 'Pond', size=18, ha='center', color='blue')
plt.text(275, 245, 'Corridor:\nWoodland', size=18, ha='center', color='brown')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.tight_layout();
plt.savefig('landscape.png', dpi=300)
```
