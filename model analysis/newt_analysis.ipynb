```python
hatchlings = np.arange(0,10)

prob = scy.poisson(2.5).pmf(hatchlings)

plt.plot(hatchlings, prob, 'bo')
plt.vlines(hatchlings, 0, prob, colors='b', lw=5, alpha=0.5)
plt.xlabel('Number of fertile female juveniles per female newt', size=12)
plt.ylabel('Probability', size=12)
plt.title('Offspring', size=14)
plt.savefig('offspring.svg');
```

```python
scenarios = pd.read_csv('scenarios.csv', skiprows = np.arange(6))
scenarios
```

```python
means_buffer = scenarios[scenarios['scenario'] == 'buffers'].groupby(['mortality-decrease-with-buffer','movement-energy']).mean()
means_corridors = scenarios[scenarios['scenario'] == 'corridors'].groupby(['mortality-decrease-with-buffer','movement-energy']).mean()

mean_values_buffer = means_buffer['total-newts'].values
mean_values_corridors = means_corridors['total-newts'].values
```

```python
heat_data_buffer = pd.DataFrame(np.reshape(mean_values_buffer, (11, 19)), 
                         index = np.unique(scenarios['mortality-decrease-with-buffer']), 
                         columns = np.unique(scenarios['movement-energy']) )

heat_data_corridors = pd.DataFrame(np.reshape(mean_values_corridors, (11, 19)), 
                         index = np.unique(scenarios['mortality-decrease-with-buffer']), 
                         columns = np.unique(scenarios['movement-energy']) )
```

```python
plt.figure(figsize= (15,8))
sns.heatmap(heat_data_buffer, annot=False, square = False)
plt.xlabel('Movement energy', size=15)
plt.ylabel('mortality decrease with buffers', size=15)
plt.show();

plt.figure(figsize= (15,8))
sns.heatmap(heat_data_corridors, annot=False, square = False)
plt.xlabel('Movement energy', size=15)
plt.ylabel('mortality decrease with buffers', size=15)
plt.show();
```

```python
plt.figure(figsize= (15,8))
sns.heatmap(heat_data_buffer - heat_data_corridors, annot=False, square = False, center = 0, cmap="RdYlGn")
plt.xlabel('Movement energy', size=15)
plt.ylabel('mortality decrease with buffers', size=15)
plt.savefig('model.svg')
plt.show();
```

```python
scenarios = pd.read_csv('scenario_new1.csv', skiprows = np.arange(6))
scenarios

means_buffer = scenarios[scenarios['scenario'] == 'buffers'].groupby(['mortality-decrease-with-buffer','movement-energy']).mean()
means_corridors = scenarios[scenarios['scenario'] == 'corridors'].groupby(['mortality-decrease-with-buffer','movement-energy']).mean()

mean_values_buffer = means_buffer['total-newts'].values
mean_values_corridors = means_corridors['total-newts'].values

heat_data_buffer = pd.DataFrame(np.reshape(mean_values_buffer, (5, 5)), 
                         index = np.unique(scenarios['mortality-decrease-with-buffer']), 
                         columns = np.unique(scenarios['movement-energy']) )

heat_data_corridors = pd.DataFrame(np.reshape(mean_values_corridors, (5, 5)), 
                         index = np.unique(scenarios['mortality-decrease-with-buffer']), 
                         columns = np.unique(scenarios['movement-energy']) )


plt.figure(figsize= (15,8))
sns.heatmap(heat_data_buffer - heat_data_corridors, annot=False, square = False, center = 0, cmap="RdYlGn")
plt.xlabel('Movement energy', size=15)
plt.ylabel('mortality decrease with buffers', size=15)
plt.savefig('model1.svg')
plt.show();
```
